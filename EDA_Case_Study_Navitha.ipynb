# Identify Missing data 

```python
# Find the columns with null values more than 50% in application data
app_missing_data = app_data.isnull().sum()
app_missing_data=app_missing_data[app_missing_data.values>(0.5*len(app_missing_data))]
len(app_missing_data)
#So there are 64 columns with null values greater than 50%
```

```python
#Drop the above 64 columns

app_missing_data = list(app_missing_data[app_missing_data.values>=0.5].index)
app_data.drop(labels=app_missing_data,axis=1,inplace=True)
print(len(app_missing_data))
```

```python
# Find the columns with null values more than 50% in previous data
prev_missing_data = prev_data.isnull().sum()
prev_missing_data=prev_missing_data[prev_missing_data.values>(0.5*len(prev_missing_data))]
len(prev_missing_data)
```

```python
#Drop the above 15 columns from previous_data

prev_missing_data = list(prev_missing_data[prev_missing_data.values>=0.5].index)
prev_data.drop(labels=prev_missing_data,axis=1,inplace=True)
print(len(prev_missing_data))
```

```python
# Removing the column values of 'XNA' and 'XAP'

prev_data=prev_data.drop(prev_data[prev_data['NAME_CASH_LOAN_PURPOSE']=='XNA'].index)
prev_data=prev_data.drop(prev_data[prev_data['NAME_CASH_LOAN_PURPOSE']=='XNA'].index)
prev_data=prev_data.drop(prev_data[prev_data['NAME_CASH_LOAN_PURPOSE']=='XAP'].index)

prev_data.shape
```

###### Now, Creating bins for continous variable categories column 'AMT_INCOME_TOTAL' and 'AMT_CREDIT'

```python
# Creating bins for income amount

bins = [0,25000,50000,75000,100000,125000,150000,175000,200000,225000,250000,275000,300000,325000,350000,375000,400000,425000,450000,475000,500000,10000000000]
slot = ['0-25000', '25000-50000','50000-75000','75000,100000','100000-125000', '125000-150000', '150000-175000','175000-200000',
       '200000-225000','225000-250000','250000-275000','275000-300000','300000-325000','325000-350000','350000-375000',
       '375000-400000','400000-425000','425000-450000','450000-475000','475000-500000','500000 and above']

app_data['AMT_INCOME_RANGE']=pd.cut(app_data['AMT_INCOME_TOTAL'],bins,labels=slot)
```

```python
# Creating bins for Credit amount

bins = [0,150000,200000,250000,300000,350000,400000,450000,500000,550000,600000,650000,700000,750000,800000,850000,900000,1000000000]
slots = ['0-150000', '150000-200000','200000-250000', '250000-300000', '300000-350000', '350000-400000','400000-450000',
        '450000-500000','500000-550000','550000-600000','600000-650000','650000-700000','700000-750000','750000-800000',
        '800000-850000','850000-900000','900000 and above']

app_data['AMT_CREDIT_RANGE']=pd.cut(app_data['AMT_CREDIT'],bins=bins,labels=slots)
```

# Find Corelation for Numerical Values for target 0 and target 1

```python

# Finding some correlation for numerical columns for both target 0 and 1 

target0_corr=target0_app_data.iloc[0:,2:]
target1_corr=target1_app_data.iloc[0:,2:]

target0=target0_corr.corr(method='spearman')
target1=target1_corr.corr(method='spearman')
```

```python
target0
```

```python
target1
```

```python
# Now, plotting the above correlation with heat map 
def targets_corr(data,title):
    plt.figure(figsize=(15, 10))
    plt.rcParams['axes.titlesize'] = 25
    plt.rcParams['axes.titlepad'] = 70
    sns.heatmap(data, cmap="RdYlGn",annot=False)

    plt.title(title)
    plt.yticks(rotation=0)
    plt.show()
```

```python
targets_corr(data=target0,title='Correlation for Non-Defaulters')
```

###### As we can see from above correlation heatmap, There are number of observation we can point out

1. less children client have in densely populated area.

# Finding any Outliers

```python
# Box plotting for univariate variables analysis in logarithmic scale
def numericalvar(var):
    sns.set_style('whitegrid')
    sns.set_context('talk')
    plt.figure(figsize=(15,5))
    plt.subplot(1, 2, 1)
    plt.yscale('log')
    sns.boxplot(x=var, data=target0_app_data,orient='v')
    plt.title('Distribution of '+ '%s' %var +' for Non-Defaulters', fontsize=14)
    plt.ylabel('Number of cases for non-defaulters')
    plt.subplot(1, 2, 2)
    plt.yscale('log')
    sns.boxplot(x= var, data=target1_app_data,orient='v')
    plt.title('Distribution of '+ '%s' %var +' for Defaulters', fontsize=14)
    plt.ylabel('Number of cases for defaulters')
    plt.show()
```

# Bivariate analysis for numerical variables

```python
def plotbivarcont(var1,var2):
    plt.figure(figsize=(16,12))
    
    plt.yscale('log')
    plt.subplot(1, 2, 1)
    sns.boxplot(data =target0_app_data, x=var1,y=var2,orient='v',hue ='NAME_FAMILY_STATUS')
    plt.title('Credit amount vs Education Status for Non-Defaulters',fontsize=10)
    plt.xticks(rotation=90)
    plt.xlabel(var1)
    plt.subplot(1, 2, 2)
    sns.boxplot(data =target1_app_data, x=var1,y=var2,orient='v',hue ='NAME_FAMILY_STATUS')
    plt.title('Credit amount vs Education Status for Defaulters',fontsize=10)
    plt.xticks(rotation=90)
    plt.xlabel(var1)
    plt.show()
```

```python
plotbivarcont('NAME_EDUCATION_TYPE','AMT_CREDIT')
```

###### Points to be noted
1. From the above box plot we can conclude that Family status of 'civil marriage', 'marriage' and 'separated' of Academic degree education are having higher number of credits than others.
2. Higher education of family status of 'marriage', 'single' and 'civil marriage' are having more outliers.
3. Civil marriage for Academic degree is having most of the credits in the third quartile.

```python
plt.figure(figsize=(16,12))
plt.subplot(1, 2, 1)
plt.yscale('log')
sns.boxplot(data =target0_app_data, x='NAME_EDUCATION_TYPE',y='AMT_INCOME_TOTAL',orient='v',hue ='NAME_FAMILY_STATUS')
plt.title('Income amount vs Education Status for Non-Defaulters',fontsize=10)
plt.xticks(rotation=90)
plt.xlabel('NAME_EDUCATION_TYPE')
plt.subplot(1, 2, 2)
plt.yscale('log')
sns.boxplot(data =target1_app_data, x='NAME_EDUCATION_TYPE',y='AMT_INCOME_TOTAL',orient='v',hue ='NAME_FAMILY_STATUS')
plt.title('Income amount vs Education Status for Defaulters',fontsize=10)
plt.xticks(rotation=90)
plt.xlabel('NAME_EDUCATION_TYPE')
plt.show()
```
